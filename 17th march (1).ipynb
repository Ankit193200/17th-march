# Q1: What are missing values in a dataset? Why is it essential to handle missing values? Name some algorithms that are not affected by missing values.

Missing values: Missing values in a dataset refer to the absence of data for certain observations or features.

Importance of handling missing values:

Missing values can lead to biased or inaccurate results in data analysis.
They can impact the performance of machine learning models.
Imputation of missing values helps in maintaining data integrity.
Algorithms not affected by missing values:

Decision Trees
Random Forests
# Q2: List down techniques used to handle missing data. Give an example of each with Python code.

Techniques:

Deletion: Removing rows or columns with missing values.
Imputation: Replacing missing values with estimated values.
Examples:

python
Copy code
import pandas as pd
import numpy as np

     # Create a sample DataFrame with missing values
data = {'A': [1, 2, np.nan, 4, 5],
        'B': [np.nan, 2, 3, 4, 5]}
df = pd.DataFrame(data)

     # Technique 1: Deletion - Drop rows with missing values
df_dropped = df.dropna()
print("DataFrame after deletion:")
print(df_dropped)

     # Technique 2: Imputation - Fill missing values with mean
df_imputed = df.fillna(df.mean())
print("\nDataFrame after imputation:")
print(df_imputed)
# Q3: Explain the imbalanced data. What will happen if imbalanced data is not handled?

Imbalanced data: Imbalanced data occurs when the distribution of classes in a dataset is not equal.

Impact of not handling imbalanced data:

Biased model predictions towards the majority class.
Poor performance on the minority class.
Difficulty in detecting the minority class.
# Q4: What are Up-sampling and Down-sampling? Explain with an example when up-sampling and down-sampling are required.

Up-sampling: Increasing the number of instances in the minority class.

Down-sampling: Reducing the number of instances in the majority class.

Example:

Up-sampling: In fraud detection, where fraud occurrences are rare.
Down-sampling: In sentiment analysis, where most reviews are positive.
# Q5: What is data Augmentation? Explain SMOTE.

Data Augmentation: Increasing the size of a dataset by applying various transformations to existing data.

SMOTE (Synthetic Minority Over-sampling Technique): A technique for up-sampling the minority class by generating synthetic examples.

# Q6: What are outliers in a dataset? Why is it essential to handle outliers?

Importance of handling outliers:

They may affect the performance and accuracy of machine learning models.
